```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
```

```python
y=np.sin(x)
z=np.cos(y)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title("sine wave")
```

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y,"r+")
plt.show()
```

```python
plt.plot(x,y,"g.")
```

```python
plt.plot(x,y,'rx')
```

```python
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),"g-")
plt.plot(x,np.cos(x),"r--")
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=["E","S","L","F","G"]
people=[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel("languages")
plt.ylabel("number of people")
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=["E","S","L","F","G"]
people=[100,50,150,40,70]
ax.pie(people,labels=languages,autopct="%1.1f%%")
plt.show()
```

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes(projection="3d")
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z)
plt.show()
```
